```python
import pandas as pd
from glob import glob
import libarchive # `pip install libarchive-c` and using with libarchive 3.4.x
from io import BytesIO
```

```python
total_size = 0
header = True
for fn in sorted(glob('*.7z')):
    print(fn)
    with libarchive.file_reader(fn) as e:
        for entry in e:
            if not entry.isfile:
                continue
            total_size += entry.size
            print(entry.name, total_size / (1024*1024))
            with BytesIO() as f:
                for i, block in enumerate(entry.get_blocks()):
                    f.write(block)
                f.seek(0)
                #df = pd.read_csv(f, nrows=10, delimiter='\t', header=None, usecols=[2, 4, 19, 20])
                df = pd.read_csv(f, delimiter='\t', header=None, usecols=[2, 4, 19, 20])
            df.columns = ['last_name', 'first_name', 'gender', 'race']
            df.to_csv('/opt/data/fl_voter/fl_voter_name_gender_race.csv', mode='a', header=header, index=False)
            header = False
```
